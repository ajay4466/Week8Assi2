```python
def reindex(df):
  new_index = pd.Index(range(1,2*len(df)+1,2))
  df.index= new_index
  return df


data = {
    'A': [1,2,3],
    'B': [4,5,6],
    'C': [7,8,9]
}

df=pd.DataFrame(data)
df_reindexed = reindex(df)
print(df_reindexed)
```

```python
data = {
    'Values' : [10,20,30,40,50]
}
df = pd.DataFrame(data)

def iterssum(df):
  sum = df['Values'].iloc[0:3].sum()
  return sum

iterssum(df)
```

```python
data = {
    'Email' : ['cr7@domain.com','messi@domain.com','john@domain.com']
}
df =pd.DataFrame(data)

def extract_username(df):
  df['UserName'] = df['Email'].str.split('@').str[0]
  return df

df = extract_username(df)

print(df)
```

```python
data = {
    'A' : [3,8,6,2,9],
    'B' : [5,2,9,3,1],
    'C' : [1,7,4,5,2]
}
df = pd.DataFrame(data)

def extract(df):
  Rows = df[(df['A']>5) & (df['B']<10)]
  return Rows

Rows=extract(df)
print(Rows)
```

```python
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

def stats(df):
  mean = df['Values'].mean()
  median = df['Values'].median()
  standard_deviation = df['Values'].std()
  return mean,median,standard_deviation

mean,median,standard_deviation = stats(df)

print(mean)
print(median)
print(standard_deviation)
df
```

```python

data = {
    'Date': pd.date_range(start='2024-10-08', periods=10, freq='D'),
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]
}
df = pd.DataFrame(data)


def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

df = calculate_moving_average(df)

print(df)
```
